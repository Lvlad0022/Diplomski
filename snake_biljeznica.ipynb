```python
def conv2d_output_size(input_size, kernel_size, stride=1, padding=0, dilation=1):
    
    return (input_size + 2*padding - dilation*(kernel_size - 1) - 1) // stride + 1

def conv2d_output_shape(H_in, W_in, kernel_size, stride=1, padding=0, dilation=1, channels_out=None):
    H_out = conv2d_output_size(H_in, kernel_size, stride, padding, dilation)
    W_out = conv2d_output_size(W_in, kernel_size, stride, padding, dilation)

    if channels_out is None:
        return H_out, W_out
    else:
        return (channels_out, H_out, W_out)
```
